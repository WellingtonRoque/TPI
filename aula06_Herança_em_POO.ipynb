# 🧬 O que é Herança?

Herança permite que uma classe reutilize características de outra classe.

Exemplo:

```python
Pessoa
    ├── Cliente
    ├── Funcionario
    └── Fornecedor
```

👉 Todas compartilham características comuns.


# 🧱 Criando a Classe Base


````python
class Pessoa:

    def __init__(self, nome, telefone, email):

        self.__nome = nome
        self.__telefone = telefone
        self.__email = email

    def exibir_dados(self):

        print(f"Nome: {self.__nome}")
        print(f"Telefone: {self.__telefone}")
        print(f"Email: {self.__email}")
````

# 🔍 Análise

Agora criamos uma classe genérica chamada:

# Pessoa

Ela servirá como base para outras classes.


# 🚀 Criando Classe Filha

Agora vamos criar a classe Cliente herdando Pessoa.


````python
class Pessoa:

    def __init__(self, nome, telefone, email):

        self.__nome = nome
        self.__telefone = telefone
        self.__email = email

    def exibir_dados(self):

        print(f"Nome: {self.__nome}")
        print(f"Telefone: {self.__telefone}")
        print(f"Email: {self.__email}")


class Cliente(Pessoa):

    pass
````

# 🚀 Testando a Herança


````python
cliente = Cliente(
    "Carlos",
    "199999999",
    "carlos@email.com"
)

cliente.exibir_dados()
````

# 🔥 O que aconteceu?

Mesmo sem escrever novamente:

- nome
- telefone
- email
- exibir_dados()

👉 a classe Cliente herdou tudo da classe Pessoa.


# ⚙️ Utilizando super()

O método:

```python
super()
```

permite acessar recursos da classe pai.


````python
class Pessoa:

    def __init__(self, nome, telefone):

        self.nome = nome
        self.telefone = telefone


class Funcionario(Pessoa):

    def __init__(self, nome, telefone, salario):

        super().__init__(nome, telefone)

        self.salario = salario
````

# 🔍 O que aconteceu?

Observe:

```python
super().__init__(nome, telefone)
```

👉 Isso executa o construtor da classe Pessoa.

Depois adicionamos:

```python
self.salario = salario
```

👉 específico do funcionário.


# 🚀 Testando Funcionario


````python
funcionario = Funcionario(
    "Maria",
    "19988887777",
    5000
)

print(funcionario.nome)
print(funcionario.telefone)
print(funcionario.salario)
````

# 🔥 Sobrescrita de Métodos

As classes filhas também podem alterar comportamentos da classe pai.


````python
class Pessoa:

    def exibir_dados(self):

        print("Exibindo dados da pessoa")


class Cliente(Pessoa):

    def exibir_dados(self):

        print("Exibindo dados do cliente")
````

# 🚀 Testando Sobrescrita


````python
cliente = Cliente()

cliente.exibir_dados()
````

# 🔍 Análise

A classe filha substituiu o comportamento da classe pai.

👉 Isso é sobrescrita de métodos.


# 🛒 Aplicando Herança no Projeto do Semestre


````python
class Pessoa:

    def __init__(self, nome, telefone, email):

        self.nome = nome
        self.telefone = telefone
        self.email = email

    def exibir_dados(self):

        print(f"Nome: {self.nome}")
        print(f"Telefone: {self.telefone}")
        print(f"Email: {self.email}")


class Cliente(Pessoa):

    def __init__(self, nome, telefone, email, limite_credito):

        super().__init__(nome, telefone, email)

        self.limite_credito = limite_credito

    def exibir_dados(self):

        print("\n===== CLIENTE =====")

        super().exibir_dados()

        print(f"Limite: R$ {self.limite_credito}")


class Funcionario(Pessoa):

    def __init__(self, nome, telefone, email, salario):

        super().__init__(nome, telefone, email)

        self.salario = salario

    def exibir_dados(self):

        print("\n===== FUNCIONÁRIO =====")

        super().exibir_dados()

        print(f"Salário: R$ {self.salario}")
````

# 🚀 Testando o Sistema


````python
cliente = Cliente(
    "Carlos",
    "199999999",
    "carlos@email.com",
    10000
)

funcionario = Funcionario(
    "Maria",
    "19988887777",
    "maria@email.com",
    5000
)

cliente.exibir_dados()

funcionario.exibir_dados()
````
