# 🔒 O que é Encapsulamento?

Encapsulamento é o processo de proteger os dados de um objeto.

Na prática isso significa:

✅ controlar alterações  
✅ validar informações  
✅ impedir modificações incorretas  
✅ organizar melhor o sistema  


# ⚠️ Classe Sem Encapsulamento


````python
class Produto:

    def __init__(self, nome, preco):

        self.nome = nome
        self.preco = preco


produto = Produto("Notebook", 3500)

print(produto.preco)

produto.preco = -1000

print(produto.preco)
````

# 🔍 Problema

Observe que o sistema permitiu:

```python
produto.preco = -1000
```

👉 Isso não deveria acontecer.

Precisamos proteger os dados.


# 🔒 Atributos Privados

Em Python utilizamos:

```python
__atributo
```

para indicar que um atributo não deve ser acessado diretamente.


````python
class Produto:

    def __init__(self, nome, preco):

        self.__nome = nome
        self.__preco = preco
````

# 📥 Getters

Getters são métodos utilizados para acessar atributos protegidos.


````python
class Produto:

    def __init__(self, nome, preco):

        self.__nome = nome
        self.__preco = preco

    def get_nome(self):

        return self.__nome

    def get_preco(self):

        return self.__preco


produto = Produto("Mouse", 120)

print(produto.get_nome())
print(produto.get_preco())
````

# 📤 Setters

Setters são métodos utilizados para alterar atributos protegidos.


````python
class Produto:

    def __init__(self, nome, preco):

        self.__nome = nome
        self.__preco = preco

    def get_nome(self):

        return self.__nome

    def get_preco(self):

        return self.__preco

    def set_preco(self, novo_preco):

        self.__preco = novo_preco


produto = Produto("Teclado", 200)

produto.set_preco(300)

print(produto.get_preco())
````

# 🚀 Adicionando Validações

Agora iremos impedir preços negativos.


````python
class Produto:

    def __init__(self, nome, preco):

        self.__nome = nome
        self.__preco = preco

    def get_nome(self):

        return self.__nome

    def get_preco(self):

        return self.__preco

    def set_preco(self, novo_preco):

        if novo_preco > 0:

            self.__preco = novo_preco

        else:

            print("Preço inválido")


produto = Produto("Monitor", 900)

produto.set_preco(-500)

print(produto.get_preco())
````

# 🔍 O que aconteceu?

Agora o sistema não permite:

❌ preços negativos  

👉 O próprio objeto controla seus dados.


# 🧱 Encapsulando Estoque


````python
class Produto:

    def __init__(self, nome, preco, estoque):

        self.__nome = nome
        self.__preco = preco
        self.__estoque = estoque

    def exibir_dados(self):

        print(f"Produto: {self.__nome}")
        print(f"Preço: R$ {self.__preco}")
        print(f"Estoque: {self.__estoque}")

    def atualizar_preco(self, novo_preco):

        if novo_preco > 0:

            self.__preco = novo_preco

        else:

            print("Preço inválido")

    def atualizar_estoque(self, novo_estoque):

        if novo_estoque >= 0:

            self.__estoque = novo_estoque

        else:

            print("Estoque inválido")
````

# 🚀 Testando o Sistema


````python
produto = Produto("Notebook", 3500, 10)

produto.exibir_dados()

print("\nAtualizando informações...\n")

produto.atualizar_preco(4000)
produto.atualizar_estoque(20)

produto.exibir_dados()

print("\nTentando inserir valores inválidos...\n")

produto.atualizar_preco(-500)
produto.atualizar_estoque(-10)
````

# 🛒 Aplicando Encapsulamento no Projeto do Semestre


````python
class Produto:

    def __init__(self, nome, preco, estoque):

        self.__nome = nome
        self.__preco = preco
        self.__estoque = estoque

    def exibir_dados(self):

        print(f"Produto: {self.__nome}")
        print(f"Preço: R$ {self.__preco}")
        print(f"Estoque: {self.__estoque}")

    def atualizar_preco(self, novo_preco):

        if novo_preco > 0:

            self.__preco = novo_preco

        else:

            print("Preço inválido")

    def atualizar_estoque(self, novo_estoque):

        if novo_estoque >= 0:

            self.__estoque = novo_estoque

        else:

            print("Estoque inválido")

    def get_nome(self):

        return self.__nome


produtos = []

while True:

    print("\n===== SISTEMA DE GESTÃO =====")
    print("1 - Cadastrar produto")
    print("2 - Listar produtos")
    print("3 - Atualizar preço")
    print("4 - Atualizar estoque")
    print("5 - Buscar produto")
    print("6 - Sair")

    opcao = input("Escolha: ")

    if opcao == "1":

        nome = input("Nome: ")
        preco = float(input("Preço: "))
        estoque = int(input("Estoque: "))

        produto = Produto(nome, preco, estoque)

        produtos.append(produto)

    elif opcao == "2":

        for p in produtos:

            p.exibir_dados()

            print("----------------")

    elif opcao == "3":

        busca = input("Produto: ")

        for p in produtos:

            if p.get_nome() == busca:

                novo = float(input("Novo preço: "))

                p.atualizar_preco(novo)

    elif opcao == "4":

        busca = input("Produto: ")

        for p in produtos:

            if p.get_nome() == busca:

                novo = int(input("Novo estoque: "))

                p.atualizar_estoque(novo)

    elif opcao == "5":

        busca = input("Buscar produto: ")

        for p in produtos:

            if p.get_nome() == busca:

                print("Produto encontrado!")

                p.exibir_dados()

    elif opcao == "6":
        break
````
