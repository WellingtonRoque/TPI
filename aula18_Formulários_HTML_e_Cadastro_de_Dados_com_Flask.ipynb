# 🚀 Criando Primeiro Formulário


## 📄 templates/cadastro_produto.html


````python
<html>

<head>

    <title>Cadastro de Produto</title>

</head>

<body>

    <h1>Cadastrar Produto</h1>

    <form method="POST">

        <label>Nome:</label>
        <input type="text" name="nome">

        <br><br>

        <label>Preço:</label>
        <input type="number" step="0.01" name="preco">

        <br><br>

        <button type="submit">
            Cadastrar
        </button>

    </form>

</body>

</html>
````

# 🔍 Explicação

## form

Define o formulário.

---

## method="POST"

Envia dados ao servidor.

---

## input

Campos de entrada.

---

## name

Nome da variável enviada ao Flask.


# 🚀 Criando Rota Flask


````python
from flask import Flask, render_template

app = Flask(__name__)


@app.route("/cadastro")
def cadastro():

    return render_template("cadastro_produto.html")


if __name__ == "__main__":

    app.run(debug=True)
````

# ⚙️ Capturando Dados do Formulário


````python
from flask import Flask, render_template, request

app = Flask(__name__)


@app.route("/cadastro", methods=["GET", "POST"])
def cadastro():

    if request.method == "POST":

        nome = request.form["nome"]

        preco = request.form["preco"]

        return f"Produto: {nome} - R$ {preco}"

    return render_template("cadastro_produto.html")
````

# 🔍 Explicação

## request.form

Captura os dados enviados pelo formulário.

---

## methods=["GET", "POST"]

Permite:

✅ exibir formulário  
✅ receber dados  


# 🌐 Diferença entre GET e POST

| Método | Função |
|---|---|
| GET | Solicita dados |
| POST | Envia dados |

👉 formulários normalmente utilizam POST.


# 🧱 Integrando ao MVC


## 📄 models/produto_model.py


````python
class Produto:

    def __init__(self, nome, preco):

        self.nome = nome
        self.preco = preco
````

## 📄 controllers/produto_controller.py


````python
from models.produto_model import Produto

produtos = []


def cadastrar_produto(nome, preco):

    produto = Produto(nome, preco)

    produtos.append(produto)


def listar_produtos():

    return produtos
````

# 🚀 Integrando Flask ao Controller


## 📄 app.py


````python
from flask import Flask, render_template, request
from controllers.produto_controller import cadastrar_produto

app = Flask(__name__)


@app.route("/cadastro", methods=["GET", "POST"])
def cadastro():

    if request.method == "POST":

        nome = request.form["nome"]

        preco = float(request.form["preco"])

        cadastrar_produto(nome, preco)

        return "Produto cadastrado!"

    return render_template("cadastro_produto.html")


if __name__ == "__main__":

    app.run(debug=True)
````

# 🌐 Criando Página de Listagem


## 📄 templates/produtos.html


````python
<html>

<head>

    <title>Produtos</title>

</head>

<body>

    <h1>Lista de Produtos</h1>

    <table border="1">

        <tr>
            <th>Nome</th>
            <th>Preço</th>
        </tr>

        {% for produto in produtos %}

        <tr>

            <td>{{ produto.nome }}</td>
            <td>R$ {{ produto.preco }}</td>

        </tr>

        {% endfor %}

    </table>

</body>

</html>
````

# 🚀 Criando Rota de Listagem


````python
from controllers.produto_controller import listar_produtos


@app.route("/produtos")
def produtos():

    lista = listar_produtos()

    return render_template(
        "produtos.html",
        produtos=lista
    )
````

# ⚠️ Problema Comum

Se o usuário digitar:

```plaintext
abc
```

no preço:

💥 ValueError

👉 precisamos validar os dados.


# 🚀 Tratando Erros no Flask


````python
from flask import Flask, render_template, request

app = Flask(__name__)


@app.route("/cadastro", methods=["GET", "POST"])
def cadastro():

    erro = ""

    if request.method == "POST":

        try:

            nome = request.form["nome"]

            preco = float(request.form["preco"])

            return f"{nome} - {preco}"

        except ValueError:

            erro = "Preço inválido"

    return render_template(
        "cadastro_produto.html",
        erro=erro
    )
````
