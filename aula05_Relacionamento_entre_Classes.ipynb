# 🧠 Contexto

Até o momento trabalhamos com classes isoladas.

Por exemplo:

- Produto

Entretanto, em sistemas reais as entidades normalmente se relacionam.

Exemplo:

👉 um Pedido possui:
- cliente
- produtos
- valor total

👉 um Cliente pode possuir:
- vários pedidos

👉 um Produto pode:

💥 Isso é relacionamento entre classes.


# 🧱 Criando a Classe Cliente


````python
class Cliente:

    def __init__(self, nome, telefone):

        self.__nome = nome
        self.__telefone = telefone

    def __str__(self):

        return f"{self.__nome} | {self.__telefone}"

    def get_nome(self):

        return self.__nome
````

# 🧱 Criando a Classe Produto


````python
class Produto:

    def __init__(self, nome, preco):

        self.__nome = nome
        self.__preco = preco

    def __str__(self):

        return f"{self.__nome} - R$ {self.__preco}"

    def get_preco(self):

        return self.__preco
````

# 🧾 Criando a Classe Pedido

Agora iremos criar a classe mais importante do sistema até o momento.

👉 O pedido irá relacionar:

- cliente
- produtos


````python
class Pedido:

    def __init__(self, cliente):

        self.__cliente = cliente
        self.__produtos = []
````

# 🔍 O que aconteceu?

Observe:

```python
self.__cliente = cliente
```

👉 Isso significa que o Pedido armazenará um OBJETO Cliente.

E:

```python
self.__produtos = []
```

👉 armazenará vários objetos Produto.


# 🚀 Adicionando Produtos ao Pedido


````python
class Pedido:

    def __init__(self, cliente):

        self.__cliente = cliente
        self.__produtos = []

    def adicionar_produto(self, produto):

        self.__produtos.append(produto)
````

# 💰 Calculando Valor Total do Pedido


````python
class Pedido:

    def __init__(self, cliente):

        self.__cliente = cliente
        self.__produtos = []

    def adicionar_produto(self, produto):

        self.__produtos.append(produto)

    def calcular_total(self):

        total = 0

        for p in self.__produtos:

            total += p.get_preco()

        return total
````

# 🔍 Análise

Observe:

```python
p.get_preco()
```

👉 Estamos utilizando métodos da classe Produto dentro da classe Pedido.

💥 Isso é relacionamento entre objetos.


# 🧾 Exibindo Dados do Pedido


````python
class Pedido:

    def __init__(self, cliente):

        self.__cliente = cliente
        self.__produtos = []

    def adicionar_produto(self, produto):

        self.__produtos.append(produto)

    def calcular_total(self):

        total = 0

        for p in self.__produtos:

            total += p.get_preco()

        return total

    def exibir_pedido(self):

        print("\n===== PEDIDO =====")

        print(f"Cliente: {self.__cliente}")

        print("\nProdutos:")

        for p in self.__produtos:

            print(p)

        print(f"\nTotal: R$ {self.calcular_total()}")
````

# 🚀 Testando o Relacionamento entre Classes


````python
cliente = Cliente("Carlos", "199999999")

produto1 = Produto("Notebook", 3500)
produto2 = Produto("Mouse", 120)
produto3 = Produto("Teclado", 200)

pedido = Pedido(cliente)

pedido.adicionar_produto(produto1)
pedido.adicionar_produto(produto2)
pedido.adicionar_produto(produto3)

pedido.exibir_pedido()
````

# 🧬 Composição

Composição ocorre quando uma classe possui objetos de outra classe.

Exemplo:

```python
Pedido -> Cliente
Pedido -> Produtos
```

👉 O Pedido depende desses objetos para existir.


# 🛒 Aplicando no Projeto do Semestre

Agora vamos criar uma primeira versão do sistema de pedidos.


````python
class Cliente:

    def __init__(self, nome, telefone):

        self.__nome = nome
        self.__telefone = telefone

    def __str__(self):

        return f"{self.__nome} | {self.__telefone}"


class Produto:

    def __init__(self, nome, preco):

        self.__nome = nome
        self.__preco = preco

    def __str__(self):

        return f"{self.__nome} - R$ {self.__preco}"

    def get_preco(self):

        return self.__preco


class Pedido:

    def __init__(self, cliente):

        self.__cliente = cliente
        self.__produtos = []

    def adicionar_produto(self, produto):

        self.__produtos.append(produto)

    def calcular_total(self):

        total = 0

        for p in self.__produtos:

            total += p.get_preco()

        return total

    def exibir_pedido(self):

        print("\n===== PEDIDO =====")

        print(f"Cliente: {self.__cliente}")

        print("\nProdutos:")

        for p in self.__produtos:

            print(p)

        print(f"\nTotal: R$ {self.calcular_total()}")


cliente = Cliente("Maria", "19988887777")

produto1 = Produto("Notebook", 3500)
produto2 = Produto("Mouse", 120)

pedido = Pedido(cliente)

pedido.adicionar_produto(produto1)
pedido.adicionar_produto(produto2)

pedido.exibir_pedido()
````
