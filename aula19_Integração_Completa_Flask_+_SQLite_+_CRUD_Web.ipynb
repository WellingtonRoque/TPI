# 🚀 Criando Conexão com Banco


## 📄 database/conexao.py


````python
import sqlite3


def conectar():

    return sqlite3.connect("sistema.db")
````

# 🚀 Criando Tabela


````python
from database.conexao import conectar

conexao = conectar()

cursor = conexao.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS produtos (

    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT,
    preco REAL

)
""")

conexao.commit()

conexao.close()

print("Tabela criada!")
````

# 🧱 Criando Model


## 📄 models/produto_model.py


````python
class Produto:

    def __init__(self, nome, preco):

        self.nome = nome
        self.preco = preco
````

# 🚀 Criando Controller


## 📄 controllers/produto_controller.py


````python
from database.conexao import conectar


def cadastrar_produto(nome, preco):

    conexao = conectar()

    cursor = conexao.cursor()

    cursor.execute("""
    INSERT INTO produtos (nome, preco)
    VALUES (?, ?)
    """, (nome, preco))

    conexao.commit()

    conexao.close()


def listar_produtos():

    conexao = conectar()

    cursor = conexao.cursor()

    cursor.execute("SELECT * FROM produtos")

    produtos = cursor.fetchall()

    conexao.close()

    return produtos
````

# 🔥 O que o Controller faz?

O Controller:

✅ conecta ao banco  
✅ executa SQL  
✅ organiza regras do sistema  
✅ separa responsabilidades

# 🚀 Integrando Cadastro ao Flask


## 📄 app.py


````python
from flask import Flask, render_template, request, redirect

from controllers.produto_controller import (
    cadastrar_produto,
    listar_produtos
)

app = Flask(__name__)


@app.route("/")
def home():

    return render_template("index.html")


@app.route("/cadastro", methods=["GET", "POST"])
def cadastro():

    if request.method == "POST":

        nome = request.form["nome"]

        preco = float(request.form["preco"])

        cadastrar_produto(nome, preco)

        return redirect("/produtos")

    return render_template("cadastro_produto.html")
````

# 🔍 Explicação

## redirect()

Redireciona o usuário após o cadastro.

# 🚀 Implementando UPDATE


## 📄 controllers/produto_controller.py


````python
def buscar_produto(id):

    conexao = conectar()

    cursor = conexao.cursor()

    cursor.execute("""
    SELECT * FROM produtos
    WHERE id = ?
    """, (id,))

    produto = cursor.fetchone()

    conexao.close()

    return produto


def atualizar_produto(id, nome, preco):

    conexao = conectar()

    cursor = conexao.cursor()

    cursor.execute("""
    UPDATE produtos
    SET nome = ?, preco = ?
    WHERE id = ?
    """, (nome, preco, id))

    conexao.commit()

    conexao.close()
````

# 🚀 Rota de Edição


````python
from controllers.produto_controller import (
    buscar_produto,
    atualizar_produto
)


@app.route("/editar/<int:id>", methods=["GET", "POST"])
def editar(id):

    produto = buscar_produto(id)

    if request.method == "POST":

        nome = request.form["nome"]

        preco = float(request.form["preco"])

        atualizar_produto(id, nome, preco)

        return redirect("/produtos")

    return render_template(
        "editar_produto.html",
        produto=produto
    )
````

# 🚀 Implementando DELETE


## 📄 controllers/produto_controller.py


````python
def excluir_produto(id):

    conexao = conectar()

    cursor = conexao.cursor()

    cursor.execute("""
    DELETE FROM produtos
    WHERE id = ?
    """, (id,))

    conexao.commit()

    conexao.close()
````

# 🚀 Rota de Exclusão


````python
from controllers.produto_controller import excluir_produto


@app.route("/excluir/<int:id>")
def excluir(id):

    excluir_produto(id)

    return redirect("/produtos")
````
