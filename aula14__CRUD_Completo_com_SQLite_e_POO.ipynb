# 🚀 Criando Banco de Dados


```python
import sqlite3

conexao = sqlite3.connect("sistema.db")

cursor = conexao.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS produtos (

    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT,
    preco REAL

)
""")

conexao.commit()

print("Banco e tabela prontos!")
```

# 🔍 Explicação

## CREATE TABLE IF NOT EXISTS

Cria a tabela apenas se ela não existir.

👉 isso evita erros ao executar o sistema múltiplas vezes.


# 🧱 Criando Classe Produto


```python
class Produto:

    def __init__(self, nome, preco):

        self.nome = nome
        self.preco = preco
```

# 🚀 CREATE — Cadastrar Produtos


```python
produto = Produto("Notebook", 3500)

cursor.execute("""
INSERT INTO produtos (nome, preco)
VALUES (?, ?)
""", (produto.nome, produto.preco))

conexao.commit()

print("Produto cadastrado!")
```

# 🔍 Explicação

## INSERT INTO

Insere dados na tabela.

---

## (?, ?)

Placeholders utilizados para segurança.


# 🚀 READ — Listar Produtos


```python
cursor.execute("SELECT * FROM produtos")

produtos = cursor.fetchall()

for produto in produtos:

    print(produto)
```

# 🔍 Explicação

## fetchall()

Retorna todos os registros encontrados.


# 🚀 UPDATE — Atualizar Produto


```python
cursor.execute("""
UPDATE produtos
SET preco = 4000
WHERE nome = 'Notebook'
""")

conexao.commit()

print("Produto atualizado!")
```

# 🔍 Explicação

## UPDATE

Atualiza registros existentes.

---

## WHERE

Define qual registro será alterado.


# 🚀 DELETE — Remover Produto


```python
cursor.execute("""
DELETE FROM produtos
WHERE nome = 'Notebook'
""")

conexao.commit()

print("Produto removido!")
```

# 🔍 Explicação

## DELETE

Remove registros da tabela.

# 🧠 Organizando Melhor o Sistema

Agora iremos criar funções específicas para cada operação.


```python
import sqlite3

conexao = sqlite3.connect("sistema.db")

cursor = conexao.cursor()


def cadastrar_produto(nome, preco):

    cursor.execute("""
    INSERT INTO produtos (nome, preco)
    VALUES (?, ?)
    """, (nome, preco))

    conexao.commit()

    print("Produto cadastrado!")


def listar_produtos():

    cursor.execute("SELECT * FROM produtos")

    produtos = cursor.fetchall()

    for produto in produtos:

        print(produto)


def atualizar_produto(id, novo_preco):

    cursor.execute("""
    UPDATE produtos
    SET preco = ?
    WHERE id = ?
    """, (novo_preco, id))

    conexao.commit()

    print("Produto atualizado!")


def excluir_produto(id):

    cursor.execute("""
    DELETE FROM produtos
    WHERE id = ?
    """, (id,))

    conexao.commit()

    print("Produto removido!")
```

# 🔍 O que melhorou?

Agora o sistema possui:

✅ responsabilidades separadas  
✅ código reutilizável  
✅ arquitetura mais limpa  


# 🚀 Criando Menu do Sistema


```python
while True:

    print("\n===== SISTEMA =====")
    print("1 - Cadastrar produto")
    print("2 - Listar produtos")
    print("3 - Atualizar produto")
    print("4 - Excluir produto")
    print("5 - Sair")

    opcao = input("Escolha: ")

    if opcao == "1":

        nome = input("Nome: ")
        preco = float(input("Preço: "))

        cadastrar_produto(nome, preco)

    elif opcao == "2":

        listar_produtos()

    elif opcao == "3":

        id_produto = int(input("ID do produto: "))
        novo_preco = float(input("Novo preço: "))

        atualizar_produto(id_produto, novo_preco)

    elif opcao == "4":

        id_produto = int(input("ID do produto: "))

        excluir_produto(id_produto)

    elif opcao == "5":

        break
```
