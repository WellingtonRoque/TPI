# 🚀 Criando Nosso Primeiro Arquivo


```python
arquivo = open("produtos.txt", "w")

arquivo.write("Notebook\n")
arquivo.write("Mouse\n")
arquivo.write("Teclado\n")

arquivo.close()

print("Arquivo criado com sucesso!")
```

# 🚀 Lendo Arquivos

# 🚀 Utilizando with

Forma profissional de trabalhar com arquivos.


```python
with open("produtos.txt", "r") as arquivo:

    conteudo = arquivo.read()

    print(conteudo)
```

# 🔍 Vantagens do with

✅ fecha automaticamente o arquivo  
✅ código mais limpo  
✅ evita problemas de recursos  


# ⚙️ Salvando Dados do Sistema


```python
produtos = [
    "Notebook",
    "Mouse",
    "Teclado"
]

with open("produtos.txt", "w") as arquivo:

    for produto in produtos:

        arquivo.write(produto + "\n")

print("Produtos salvos!")
```

# 🚀 Carregando Dados do Sistema


```python
produtos = []

with open("produtos.txt", "r") as arquivo:

    for linha in arquivo:

        produtos.append(linha.strip())

print(produtos)
```

# 🔍 O que aconteceu?

Agora os dados:

✅ são salvos  
✅ podem ser recuperados  
✅ continuam após fechar o sistema  


# 🛒 Aplicando Persistência no Projeto do Semestre


```python
class Produto:

    def __init__(self, nome, preco):

        self.nome = nome
        self.preco = preco

    def __str__(self):

        return f"{self.nome};{self.preco}"
```

# 🚀 Salvando Produtos


```python
produtos = [
    Produto("Notebook", 3500),
    Produto("Mouse", 120),
    Produto("Monitor", 1500)
]

with open("produtos.txt", "w") as arquivo:

    for produto in produtos:

        arquivo.write(str(produto) + "\n")

print("Produtos salvos!")
```

# 🚀 Carregando Produtos


```python
produtos_carregados = []

with open("produtos.txt", "r") as arquivo:

    for linha in arquivo:

        dados = linha.strip().split(";")

        nome = dados[0]
        preco = float(dados[1])

        produto = Produto(nome, preco)

        produtos_carregados.append(produto)

for produto in produtos_carregados:

    print(produto.nome, produto.preco)
```

# 🔥 O que acabamos de fazer?

Agora o sistema possui:

✅ persistência  
✅ recuperação de dados  
✅ armazenamento permanente  
✅ sistema mais profissional

# 🛒 Evoluindo o Sistema Comercial


```python
class Produto:

    def __init__(self, nome, preco):

        self.nome = nome
        self.preco = preco

    def __str__(self):

        return f"{self.nome};{self.preco}"


produtos = []

while True:

    print("\n===== SISTEMA =====")
    print("1 - Cadastrar produto")
    print("2 - Listar produtos")
    print("3 - Salvar produtos")
    print("4 - Carregar produtos")
    print("5 - Sair")

    opcao = input("Escolha: ")

    if opcao == "1":

        nome = input("Nome: ")
        preco = float(input("Preço: "))

        produto = Produto(nome, preco)

        produtos.append(produto)

    elif opcao == "2":

        for produto in produtos:

            print(produto.nome, produto.preco)

    elif opcao == "3":

        with open("produtos.txt", "w") as arquivo:

            for produto in produtos:

                arquivo.write(str(produto) + "\n")

        print("Produtos salvos!")

    elif opcao == "4":

        produtos = []

        try:

            with open("produtos.txt", "r") as arquivo:

                for linha in arquivo:

                    dados = linha.strip().split(";")

                    nome = dados[0]
                    preco = float(dados[1])

                    produto = Produto(nome, preco)

                    produtos.append(produto)

            print("Produtos carregados!")

        except FileNotFoundError:

            print("Arquivo não encontrado")

    elif opcao == "5":

        break
```
