# 🧠 Contexto

Imagine o seguinte cenário:

No sistema temos:

- Cliente
- Funcionário
- Fornecedor

Todos possuem o método:

```python
exibir_dados()
```

Entretanto:

👉 cada tipo de objeto deve exibir informações diferentes.

💥 Problema:
Como utilizar o mesmo método com comportamentos diferentes?

👉 É exatamente aqui que entra o Polimorfismo.


# 🔥 O que é Polimorfismo?

A palavra polimorfismo significa:

# 🔹 muitas formas

Na Programação Orientada a Objetos isso significa:

👉 objetos diferentes podem responder de formas diferentes ao mesmo método.


# 🚀 Exemplo Simples


````python
class Animal:

    def emitir_som(self):

        print("Som genérico")


class Cachorro(Animal):

    def emitir_som(self):

        print("Au Au")


class Gato(Animal):

    def emitir_som(self):

        print("Miau")
````

# 🚀 Testando o Polimorfismo


````python
cachorro = Cachorro()
gato = Gato()

cachorro.emitir_som()
gato.emitir_som()
````

# 🔍 O que aconteceu?

Todos os objetos possuem:

```python
emitir_som()
```

Entretanto:

👉 cada objeto executa um comportamento diferente.

💥 Isso é Polimorfismo.


# 🧠 Listas Polimórficas

Agora vamos armazenar objetos diferentes em uma lista.


````python
animais = [
    Cachorro(),
    Gato(),
    Cachorro(),
    Gato()
]

for animal in animais:

    animal.emitir_som()
````

# 🔍 Análise

Observe:

👉 o loop não precisa saber qual objeto está utilizando.

Cada objeto responde da sua própria forma.


# 🛒 Aplicando Polimorfismo no Projeto do Semestre


````python
class Pessoa:

    def __init__(self, nome):

        self.nome = nome

    def exibir_dados(self):

        print(f"Pessoa: {self.nome}")


class Cliente(Pessoa):

    def exibir_dados(self):

        print(f"Cliente: {self.nome}")


class Funcionario(Pessoa):

    def exibir_dados(self):

        print(f"Funcionário: {self.nome}")


class Fornecedor(Pessoa):

    def exibir_dados(self):

        print(f"Fornecedor: {self.nome}")
````

# 🚀 Testando o Sistema


````python
cliente = Cliente("Carlos")
funcionario = Funcionario("Maria")
fornecedor = Fornecedor("Tech Informática")

cliente.exibir_dados()

funcionario.exibir_dados()

fornecedor.exibir_dados()
````

# 🔥 Utilizando Polimorfismo no Sistema


````python
usuarios = [
    Cliente("Carlos"),
    Funcionario("Maria"),
    Fornecedor("Tech Informática")
]

for usuario in usuarios:

    usuario.exibir_dados()
````

# 🔍 O que aconteceu?

O sistema conseguiu trabalhar com:

- Cliente
- Funcionario
- Fornecedor

utilizando:

```python
usuario.exibir_dados()
```

👉 sem precisar verificar o tipo do objeto.

💥 Isso deixa o sistema muito mais profissional.

# 🔹 sobrescrita de métodos

As classes filhas alteram o comportamento da classe pai.


# 💳 Exemplo Profissional — Sistema de Pagamentos


````python
class Pagamento:

    def processar_pagamento(self):

        print("Processando pagamento...")


class Pix(Pagamento):

    def processar_pagamento(self):

        print("Pagamento realizado via PIX")


class Cartao(Pagamento):

    def processar_pagamento(self):

        print("Pagamento realizado via Cartão")


class Boleto(Pagamento):

    def processar_pagamento(self):

        print("Pagamento realizado via Boleto")
````

# 🚀 Testando o Sistema de Pagamentos


````python
pagamentos = [
    Pix(),
    Cartao(),
    Boleto()
]

for pagamento in pagamentos:

    pagamento.processar_pagamento()
````

# 🔍 Análise

Observe:

👉 todos possuem o mesmo método:

```python
processar_pagamento()
```

Entretanto:

👉 cada classe executa de forma diferente.

💥 Isso é extremamente utilizado em sistemas reais.


# 🛒 Evoluindo o Sistema de Gestão Comercial

Agora vamos aplicar polimorfismo no sistema de pedidos.


````python
class Produto:

    def __init__(self, nome, preco):

        self.nome = nome
        self.preco = preco

    def calcular_desconto(self):

        return self.preco


class ProdutoPromocao(Produto):

    def calcular_desconto(self):

        return self.preco * 0.9


class ProdutoVip(Produto):

    def calcular_desconto(self):

        return self.preco * 0.8
````

# 🚀 Testando Produtos com Polimorfismo


````python
produto1 = Produto("Notebook", 3500)
produto2 = ProdutoPromocao("Mouse", 120)
produto3 = ProdutoVip("Monitor", 1500)

produtos = [produto1, produto2, produto3]

for p in produtos:

    print(f"{p.nome} -> R$ {p.calcular_desconto()}")
````
