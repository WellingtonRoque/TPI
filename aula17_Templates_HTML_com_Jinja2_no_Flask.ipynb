# ⚙️ Enviando Dados para o HTML


````python
from flask import Flask, render_template

app = Flask(__name__)

produtos = [
    "Notebook",
    "Mouse",
    "Monitor"
]


@app.route("/produtos")
def listar_produtos():

    return render_template(
        "produtos.html",
        produtos=produtos
    )
````

# 🌐 Exibindo Dados no HTML


## 📄 templates/produtos.html


````python
<html>

<head>

    <title>Produtos</title>

</head>

<body>

    <h1>Lista de Produtos</h1>

    <ul>

        {% for produto in produtos %}

            <li>{{ produto }}</li>

        {% endfor %}

    </ul>

</body>

</html>
````

# 🔍 Explicação

## {% %}

Estruturas do Jinja2.

---

## {{ }}

Exibe variáveis no HTML.


# 🔥 O que acabamos de fazer?

Agora os dados do Python aparecem dinamicamente na página web.


# ⚙️ Trabalhando com Objetos


````python
class Produto:

    def __init__(self, nome, preco):

        self.nome = nome
        self.preco = preco


produtos = [
    Produto("Notebook", 3500),
    Produto("Mouse", 120),
    Produto("Monitor", 1500)
]
````

# 🌐 Exibindo Objetos no HTML


## 📄 templates/produtos.html


````python
<table border="1">

    <tr>
        <th>Nome</th>
        <th>Preço</th>
    </tr>

    {% for produto in produtos %}

    <tr>

        <td>{{ produto.nome }}</td>
        <td>R$ {{ produto.preco }}</td>

    </tr>

    {% endfor %}

</table>
````

# 🚀 Integrando MVC ao Template


## 📄 controllers/produto_controller.py


````python
class Produto:

    def __init__(self, nome, preco):

        self.nome = nome
        self.preco = preco


produtos = [
    Produto("Notebook", 3500),
    Produto("Mouse", 120)
]


def listar_produtos():

    return produtos
````

## 📄 app.py


````python
from flask import Flask, render_template
from controllers.produto_controller import listar_produtos

app = Flask(__name__)


@app.route("/produtos")
def produtos():

    lista = listar_produtos()

    return render_template(
        "produtos.html",
        produtos=lista
    )


if __name__ == "__main__":

    app.run(debug=True)
````
