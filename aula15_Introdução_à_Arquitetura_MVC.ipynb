# 🧱 O que é MVC?

MVC significa:

| Camada | Responsabilidade |
|---|---|
| Model | Dados |
| View | Interface |
| Controller | Regras do sistema |

👉 cada parte possui uma responsabilidade específica.


# 🔍 Problema Sem MVC


````python
import sqlite3

conexao = sqlite3.connect("sistema.db")

cursor = conexao.cursor()

nome = input("Nome: ")
preco = float(input("Preço: "))

cursor.execute("""
INSERT INTO produtos (nome, preco)
VALUES (?, ?)
""", (nome, preco))

conexao.commit()

print("Produto cadastrado!")
````

# ⚠️ Problema

Nesse exemplo temos tudo misturado:

❌ interface  
❌ banco de dados  
❌ regras do sistema  

👉 isso dificulta manutenção e crescimento.

# 🔍 O que cada camada faz?

## Model

Responsável pelos dados.

👉 representa entidades do sistema.

---

## View

Responsável pela interface.

👉 interação com usuário.

---

## Controller

Responsável pelas regras do sistema.

👉 conecta View e Model.


# 🧱 Criando o Model


## 📄 produto_model.py


````python
class Produto:

    def __init__(self, nome, preco):

        self.nome = nome
        self.preco = preco
````

# 🔍 Responsabilidade do Model

O Model representa os dados do sistema.

👉 normalmente contém:

- atributos
- validações
- regras básicas


# 🧱 Criando a View


## 📄 produto_view.py


````python
def menu():

    print("\n===== SISTEMA =====")
    print("1 - Cadastrar produto")
    print("2 - Listar produtos")
    print("3 - Sair")


def obter_dados_produto():

    nome = input("Nome: ")
    preco = float(input("Preço: "))

    return nome, preco
````

# 🔍 Responsabilidade da View

A View é responsável pela interface.

👉 ela apenas:

- exibe informações
- coleta dados do usuário


# 🧱 Criando o Controller


## 📄 produto_controller.py


````python
from produto_model import Produto

produtos = []


def cadastrar_produto(nome, preco):

    produto = Produto(nome, preco)

    produtos.append(produto)

    print("Produto cadastrado!")


def listar_produtos():

    for produto in produtos:

        print(produto.nome, produto.preco)
````

# 🔍 Responsabilidade do Controller

O Controller:

✅ recebe dados da View  
✅ aplica regras  
✅ manipula Models  
✅ controla o fluxo do sistema  


# 🚀 Criando o Sistema Principal


## 📄 app.py


````python
from produto_view import menu, obter_dados_produto
from produto_controller import cadastrar_produto, listar_produtos


while True:

    menu()

    opcao = input("Escolha: ")

    if opcao == "1":

        nome, preco = obter_dados_produto()

        cadastrar_produto(nome, preco)

    elif opcao == "2":

        listar_produtos()

    elif opcao == "3":

        break
````

# 🔥 O que acabamos de construir?

Agora o sistema possui:

✅ separação de responsabilidades  
✅ arquitetura profissional  
✅ código organizado  
✅ manutenção simplificada  


# 🧠 Fluxo do MVC

👉 Usuário interage com:

# View

↓

👉 View envia dados para:

# Controller

↓

👉 Controller manipula:

# Model

↓

👉 Resultado retorna para:

# View


# 🛒 Aplicando MVC com SQLite


## 📄 conexao.py


````python
import sqlite3


def conectar():

    return sqlite3.connect("sistema.db")
````

## 📄 produto_controller.py


````python
from database.conexao import conectar


def listar_produtos():

    conexao = conectar()

    cursor = conexao.cursor()

    cursor.execute("SELECT * FROM produtos")

    produtos = cursor.fetchall()

    conexao.close()

    return produtos
````
