# ⚙️ O que é Sobrecarga?

Sobrecarga é a capacidade de um método aceitar diferentes formas de utilização.

Em algumas linguagens podemos criar:

```python
somar(a, b)
somar(a, b, c)
```

👉 Python trabalha isso de forma diferente utilizando:

- parâmetros opcionais
- *args
- **kwargs


# 🚀 Parâmetros Opcionais


````python
class Produto:

    def __init__(self, nome, preco=0, estoque=0):

        self.nome = nome
        self.preco = preco
        self.estoque = estoque

    def exibir_dados(self):

        print(f"Produto: {self.nome}")
        print(f"Preço: R$ {self.preco}")
        print(f"Estoque: {self.estoque}")
````

# 🚀 Criando Objetos Flexíveis


````python
produto1 = Produto("Notebook", 3500, 10)

produto2 = Produto("Mouse", 120)

produto3 = Produto("Teclado")

produto1.exibir_dados()

print()

produto2.exibir_dados()

print()

produto3.exibir_dados()
````

# 🔍 O que aconteceu?

Agora o construtor aceita diferentes quantidades de parâmetros.

👉 Isso torna o sistema mais flexível.


# ⚙️ Utilizando *args


````python
def somar(*numeros):

    total = 0

    for n in numeros:

        total += n

    return total


print(somar(10, 20))

print(somar(10, 20, 30))

print(somar(10, 20, 30, 40))
````

# 🔍 Explicação

O operador:

```python
*args
```

permite receber múltiplos valores.

👉 Isso cria métodos mais flexíveis.


# ⚙️ Utilizando **kwargs


````python
def cadastrar_produto(**dados):

    for chave, valor in dados.items():

        print(f"{chave}: {valor}")


cadastrar_produto(
    nome="Notebook",
    preco=3500,
    estoque=10
)
````

# 🔍 Explicação

O operador:

```python
**kwargs
```

permite trabalhar com parâmetros nomeados dinâmicos.


# 🛒 Aplicando no Projeto do Semestre


````python
class Produto:

    def __init__(self, nome, preco=0, estoque=0):

        self.nome = nome
        self.preco = preco
        self.estoque = estoque

    def atualizar_produto(self, preco=None, estoque=None):

        if preco is not None:

            self.preco = preco

        if estoque is not None:

            self.estoque = estoque

    def exibir_dados(self):

        print(f"Produto: {self.nome}")
        print(f"Preço: R$ {self.preco}")
        print(f"Estoque: {self.estoque}")
````

# 🚀 Testando Atualizações Flexíveis


````python
produto = Produto("Monitor", 1500, 5)

produto.exibir_dados()

print("\nAtualizando somente preço...\n")

produto.atualizar_produto(preco=1800)

produto.exibir_dados()

print("\nAtualizando estoque...\n")

produto.atualizar_produto(estoque=20)

produto.exibir_dados()
````

# 🔥 Métodos Mais Inteligentes

Agora nosso sistema consegue:

✅ atualizar apenas preço  
✅ atualizar apenas estoque  
✅ atualizar ambos  

👉 Isso reduz duplicação de código.

# 🔍 Problema

O método possui muitas responsabilidades.

👉 Isso dificulta manutenção.


# ✅ Exemplo Melhor


````python
class Produto:

    def cadastrar(self):

        print("Cadastrando produto")

    def atualizar(self):

        print("Atualizando produto")

    def excluir(self):

        print("Excluindo produto")
````

# 🔥 Organização Profissional

Em sistemas reais normalmente seguimos:

# 🔹 Uma responsabilidade por método

👉 Isso melhora muito a arquitetura.


# 🛒 Evoluindo o Sistema Comercial


````python
class Produto:

    def __init__(self, nome, preco=0, estoque=0):

        self.nome = nome
        self.preco = preco
        self.estoque = estoque

    def atualizar_preco(self, preco):

        self.preco = preco

    def atualizar_estoque(self, estoque):

        self.estoque = estoque

    def calcular_valor_total(self):

        return self.preco * self.estoque

    def exibir_dados(self):

        print(f"Produto: {self.nome}")
        print(f"Preço: R$ {self.preco}")
        print(f"Estoque: {self.estoque}")
        print(f"Valor Total: R$ {self.calcular_valor_total()}")
````

# 🚀 Testando o Sistema


````python
produto = Produto("Notebook", 3500, 10)

produto.exibir_dados()
````

# 🧠 Arquitetura Limpa

Observe como o sistema está ficando:

✅ modular  
✅ organizado  
✅ reutilizável  
✅ mais profissional
