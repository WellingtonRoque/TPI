# 🧠 Contexto

Imagine o seguinte cenário:

No sistema temos diferentes tipos de pagamento:

- PIX
- Cartão
- Boleto

Todos devem possuir:

```python
processar_pagamento()
```

Entretanto:

👉 não faz sentido criar um pagamento genérico funcionando sozinho.

Precisamos apenas obrigar as subclasses a implementarem esse comportamento.

👉 É exatamente aqui que entram as Classes Abstratas.


# 🧱 O que é uma Classe Abstrata?

Uma classe abstrata funciona como um modelo base.

Ela:

✅ define comportamentos obrigatórios  
✅ organiza a arquitetura  
✅ impede criação de objetos incompletos  
✅ melhora a reutilização  

👉 Normalmente não criamos objetos diretamente da classe abstrata.


# ⚠️ Problema Sem Classe Abstrata


````python
class Pagamento:

    def processar_pagamento(self):

        print("Processando pagamento...")


pagamento = Pagamento()

pagamento.processar_pagamento()
````

# 🔍 Problema

O sistema permitiu criar:

```python
Pagamento()
```

👉 Mas o que seria um pagamento genérico?

Não faz muito sentido.

Precisamos obrigar cada tipo de pagamento a implementar sua própria lógica.


# 🚀 Utilizando Classes Abstratas


````python
from abc import ABC, abstractmethod


class Pagamento(ABC):

    @abstractmethod
    def processar_pagamento(self):

        pass
````

# 🔍 Explicação

## ABC

Transforma a classe em abstrata.

---

## @abstractmethod

Define métodos obrigatórios para as subclasses.


# ⚠️ Tentando Criar Objeto Abstrato


````python
from abc import ABC, abstractmethod


class Pagamento(ABC):

    @abstractmethod
    def processar_pagamento(self):

        pass


# pagamento = Pagamento()
````

# 🔥 O que acontece?

Agora o Python NÃO permite:

```python
Pagamento()
```

👉 Isso protege a arquitetura do sistema.


# 🧱 Criando Subclasses


````python
from abc import ABC, abstractmethod


class Pagamento(ABC):

    @abstractmethod
    def processar_pagamento(self):

        pass


class Pix(Pagamento):

    def processar_pagamento(self):

        print("Pagamento realizado via PIX")


class Cartao(Pagamento):

    def processar_pagamento(self):

        print("Pagamento realizado via Cartão")


class Boleto(Pagamento):

    def processar_pagamento(self):

        print("Pagamento realizado via Boleto")
````

# 🚀 Testando as Subclasses


````python
pix = Pix()
cartao = Cartao()
boleto = Boleto()

pix.processar_pagamento()

cartao.processar_pagamento()

boleto.processar_pagamento()
````

# 🔍 O que aconteceu?

Agora todas as subclasses são obrigadas a implementar:

```python
processar_pagamento()
```

👉 Isso deixa a arquitetura muito mais profissional.


# 🔥 Classes Abstratas + Polimorfismo


````python
pagamentos = [
    Pix(),
    Cartao(),
    Boleto()
]

for pagamento in pagamentos:

    pagamento.processar_pagamento()
````

# 🧠 Análise

Observe:

👉 o sistema trabalha com múltiplos tipos de pagamento.

Entretanto:

👉 todos seguem o mesmo contrato definido pela classe abstrata.


# 🛒 Aplicando no Projeto do Semestre

Agora vamos aplicar classes abstratas no Sistema de Gestão Comercial.


````python
from abc import ABC, abstractmethod


class Usuario(ABC):

    def __init__(self, nome):

        self.nome = nome

    @abstractmethod
    def acessar_sistema(self):

        pass
````

# 🚀 Criando Usuários do Sistema


````python
class Administrador(Usuario):

    def acessar_sistema(self):

        print(f"{self.nome} acessou como administrador")


class Vendedor(Usuario):

    def acessar_sistema(self):

        print(f"{self.nome} acessou como vendedor")


class Estoquista(Usuario):

    def acessar_sistema(self):

        print(f"{self.nome} acessou como estoquista")
````

# 🚀 Testando o Sistema


````python
usuarios = [
    Administrador("Carlos"),
    Vendedor("Maria"),
    Estoquista("João")
]

for usuario in usuarios:

    usuario.acessar_sistema()
````
