# 🌐 O que é uma Rota?

Uma rota representa um endereço do sistema.

Exemplo:

```plaintext

```

👉 página inicial.

---

```plaintext

```

👉 página de produtos.


# 🚀 Criando Primeira Rota


````python
from flask import Flask

app = Flask(__name__)


@app.route("/")
def home():

    return "Sistema Comercial"


if __name__ == "__main__":

    app.run(debug=True)
````

# 🔍 Explicação

## @app.route("/")

Define a rota da aplicação.

---

## def home()

Função executada ao acessar a rota.

---

## return

Conteúdo exibido no navegador.

# 🌐 Criando Novas Rotas


````python
from flask import Flask

app = Flask(__name__)


@app.route("/")
def home():

    return "Página Inicial"


@app.route("/produtos")
def produtos():

    return "Página de Produtos"


@app.route("/clientes")
def clientes():

    return "Página de Clientes"


if __name__ == "__main__":

    app.run(debug=True)
````

# 🔥 O que acabamos de fazer?

Agora o sistema possui múltiplas páginas web.

# ⚙️ Rotas com Parâmetros


````python
from flask import Flask

app = Flask(__name__)


@app.route("/produto/<nome>")
def produto(nome):

    return f"Produto: {nome}"


if __name__ == "__main__":

    app.run(debug=True)
````

# 🔍 Explicação

Agora a rota recebe valores dinamicamente.

Exemplo:

```plaintext

```

# 🧱 Criando o Model


## 📄 models/produto_model.py


````python
class Produto:

    def __init__(self, nome, preco):

        self.nome = nome
        self.preco = preco
````

# 🧱 Criando o Controller


## 📄 controllers/produto_controller.py


````python
from models.produto_model import Produto

produtos = []


def cadastrar_produto(nome, preco):

    produto = Produto(nome, preco)

    produtos.append(produto)


def listar_produtos():

    return produtos
````

# 🌐 Integrando Flask ao Controller


## 📄 app.py


````python
from flask import Flask
from controllers.produto_controller import cadastrar_produto, listar_produtos

app = Flask(__name__)


@app.route("/")
def home():

    return "Sistema Comercial"


@app.route("/cadastrar")
def cadastrar():

    cadastrar_produto("Notebook", 3500)

    return "Produto cadastrado!"


@app.route("/produtos")
def produtos():

    lista = listar_produtos()

    resultado = ""

    for produto in lista:

        resultado += f"{produto.nome} - R$ {produto.preco}<br>"

    return resultado


if __name__ == "__main__":

    app.run(debug=True)
````
