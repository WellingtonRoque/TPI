# 🧱 Relembrando a Classe Produto


````python
class Produto:

    def __init__(self, nome, preco, estoque):

        self.nome = nome
        self.preco = preco
        self.estoque = estoque

    def exibir_dados(self):

        print(f"Produto: {self.nome}")
        print(f"Preço: R$ {self.preco}")
        print(f"Estoque: {self.estoque}")
````

# 🚀 Criando Múltiplos Objetos

Agora vamos criar vários produtos do sistema.


````python
produto1 = Produto("Notebook", 3500, 10)
produto2 = Produto("Mouse", 120, 50)
produto3 = Produto("Teclado", 200, 30)

produto1.exibir_dados()

print()

produto2.exibir_dados()

print()

produto3.exibir_dados()
````

# 🧠 Listas de Objetos

Em sistemas reais normalmente armazenamos objetos em listas.


````python
produtos = []

produto1 = Produto("Notebook", 3500, 10)
produto2 = Produto("Mouse", 120, 50)
produto3 = Produto("Teclado", 200, 30)

produtos.append(produto1)
produtos.append(produto2)
produtos.append(produto3)

print(produtos)
````

# 🔍 O que aconteceu?

Agora nossa lista não armazena:

- dicionários
- strings
- números

👉 Ela armazena OBJETOS.

Cada item da lista é um objeto da classe Produto.


# 🚀 Percorrendo Listas de Objetos


````python
for p in produtos:

    p.exibir_dados()

    print("----------------")
````

# 🔎 Buscando Objetos

Agora vamos buscar produtos pelo nome.


````python
busca = input("Digite o nome do produto: ")

for p in produtos:

    if p.nome == busca:

        print("Produto encontrado!")

        p.exibir_dados()
````

# 🧠 Análise Importante

Observe que agora acessamos atributos diretamente:

```python
p.nome
```

👉 Isso acontece porque p é um objeto.


# ✏️ Alterando Atributos de Objetos


````python
produto = Produto("Monitor", 900, 15)

produto.exibir_dados()

print("\nAtualizando preço...\n")

produto.preco = 1200

produto.exibir_dados()
````

# 🧱 Melhorando a Classe com Métodos

Agora vamos adicionar métodos para manipular os próprios objetos.


````python
class Produto:

    def __init__(self, nome, preco, estoque):

        self.nome = nome
        self.preco = preco
        self.estoque = estoque

    def exibir_dados(self):

        print(f"Produto: {self.nome}")
        print(f"Preço: R$ {self.preco}")
        print(f"Estoque: {self.estoque}")

    def atualizar_preco(self, novo_preco):

        self.preco = novo_preco

    def atualizar_estoque(self, novo_estoque):

        self.estoque = novo_estoque
````

# 🚀 Testando os Métodos


````python
produto = Produto("Cadeira Gamer", 1500, 5)

produto.exibir_dados()

print("\nAtualizando informações...\n")

produto.atualizar_preco(1800)
produto.atualizar_estoque(8)

produto.exibir_dados()
````

# 🛒 Aplicando no Projeto do Semestre

Agora vamos evoluir nosso sistema comercial.


````python
class Produto:

    def __init__(self, nome, preco, estoque):

        self.nome = nome
        self.preco = preco
        self.estoque = estoque

    def exibir_dados(self):

        print(f"Produto: {self.nome}")
        print(f"Preço: R$ {self.preco}")
        print(f"Estoque: {self.estoque}")

    def atualizar_preco(self, novo_preco):

        self.preco = novo_preco

    def atualizar_estoque(self, novo_estoque):

        self.estoque = novo_estoque


produtos = []

while True:

    print("\n===== SISTEMA DE GESTÃO =====")
    print("1 - Cadastrar produto")
    print("2 - Listar produtos")
    print("3 - Buscar produto")
    print("4 - Atualizar estoque")
    print("5 - Atualizar preço")
    print("6 - Sair")

    opcao = input("Escolha: ")

    if opcao == "1":

        nome = input("Nome: ")
        preco = float(input("Preço: "))
        estoque = int(input("Estoque: "))

        produto = Produto(nome, preco, estoque)

        produtos.append(produto)

    elif opcao == "2":

        for p in produtos:

            p.exibir_dados()

            print("----------------")

    elif opcao == "3":

        busca = input("Buscar produto: ")

        for p in produtos:

            if p.nome == busca:

                print("Produto encontrado!")

                p.exibir_dados()

    elif opcao == "4":

        busca = input("Produto: ")

        for p in produtos:

            if p.nome == busca:

                novo = int(input("Novo estoque: "))

                p.atualizar_estoque(novo)

    elif opcao == "5":

        busca = input("Produto: ")

        for p in produtos:

            if p.nome == busca:

                novo = float(input("Novo preço: "))

                p.atualizar_preco(novo)

    elif opcao == "6":
        break
````
