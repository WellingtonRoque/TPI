# ⚙️ O Método __init__

O método:

```python
__init__()
```

é chamado automaticamente quando criamos um objeto.

Ele é conhecido como:

# 🔹 Método Construtor

Sua principal função é inicializar os atributos do objeto.


# 🚀 Exemplo Básico de Construtor


````python
class Produto:

    def __init__(self, nome, preco):

        self.nome = nome
        self.preco = preco


produto = Produto("Notebook", 3500)

print(produto.nome)
print(produto.preco)
````

# 🔍 O que aconteceu?

Quando executamos:

```python
produto = Produto("Notebook", 3500)
```

o Python automaticamente executa:

```python
__init__()
```

👉 Isso permite configurar o objeto logo na criação.


# 🧱 Construtores Mais Organizados

Agora vamos melhorar nossa classe Produto.


````python
class Produto:

    def __init__(self, nome, preco, estoque):

        self.__nome = nome
        self.__preco = preco
        self.__estoque = estoque

    def exibir_dados(self):

        print(f"Produto: {self.__nome}")
        print(f"Preço: R$ {self.__preco}")
        print(f"Estoque: {self.__estoque}")
````

# 🔥 Método Especial __str__

O método:

```python
__str__()
```

define como o objeto será exibido ao utilizar:

```python
print(objeto)
```


# 🚀 Exemplo Sem __str__


````python
class Cliente:

    def __init__(self, nome):

        self.nome = nome


cliente = Cliente("Carlos")

print(cliente)
````

# ⚠️ Problema

O Python exibiu:

```python
<__main__.Cliente object at ...>
```

👉 Isso não é amigável.


# 🚀 Utilizando __str__


````python
class Cliente:

    def __init__(self, nome):

        self.nome = nome

    def __str__(self):

        return f"Cliente: {self.nome}"


cliente = Cliente("Carlos")

print(cliente)
````

# 🔍 O que mudou?

Agora o objeto possui uma representação textual amigável.

👉 Isso é muito utilizado em sistemas reais.


# 🛒 Aplicando __str__ no Projeto do Semestre


````python
class Produto:

    def __init__(self, nome, preco, estoque):

        self.__nome = nome
        self.__preco = preco
        self.__estoque = estoque

    def __str__(self):

        return f"{self.__nome} - R$ {self.__preco} - Estoque: {self.__estoque}"

    def atualizar_preco(self, novo_preco):

        if novo_preco > 0:

            self.__preco = novo_preco

        else:

            print("Preço inválido")

    def atualizar_estoque(self, novo_estoque):

        if novo_estoque >= 0:

            self.__estoque = novo_estoque

        else:

            print("Estoque inválido")
````

# 🚀 Testando o Método __str__


````python
produto1 = Produto("Notebook", 3500, 10)
produto2 = Produto("Mouse", 120, 50)

print(produto1)
print(produto2)
````

# 🧱 Evoluindo o Sistema Comercial

Agora vamos melhorar o sistema utilizando os métodos especiais.


````python
class Produto:

    def __init__(self, nome, preco, estoque):

        self.__nome = nome
        self.__preco = preco
        self.__estoque = estoque

    def __str__(self):

        return f"{self.__nome} | R$ {self.__preco} | Estoque: {self.__estoque}"

    def atualizar_preco(self, novo_preco):

        if novo_preco > 0:

            self.__preco = novo_preco

        else:

            print("Preço inválido")

    def atualizar_estoque(self, novo_estoque):

        if novo_estoque >= 0:

            self.__estoque = novo_estoque

        else:

            print("Estoque inválido")

    def get_nome(self):

        return self.__nome


produtos = []

while True:

    print("\n===== SISTEMA DE GESTÃO =====")
    print("1 - Cadastrar produto")
    print("2 - Listar produtos")
    print("3 - Atualizar preço")
    print("4 - Atualizar estoque")
    print("5 - Buscar produto")
    print("6 - Sair")

    opcao = input("Escolha: ")

    if opcao == "1":

        nome = input("Nome: ")
        preco = float(input("Preço: "))
        estoque = int(input("Estoque: "))

        produto = Produto(nome, preco, estoque)

        produtos.append(produto)

    elif opcao == "2":

        for p in produtos:

            print(p)

    elif opcao == "3":

        busca = input("Produto: ")

        for p in produtos:

            if p.get_nome() == busca:

                novo = float(input("Novo preço: "))

                p.atualizar_preco(novo)

    elif opcao == "4":

        busca = input("Produto: ")

        for p in produtos:

            if p.get_nome() == busca:

                novo = int(input("Novo estoque: "))

                p.atualizar_estoque(novo)

    elif opcao == "5":

        busca = input("Buscar produto: ")

        for p in produtos:

            if p.get_nome() == busca:

                print("Produto encontrado!")

                print(p)

    elif opcao == "6":
        break
````

# 🚀 Construtores Inteligentes

Os construtores também podem realizar validações.


````python
class Produto:

    def __init__(self, nome, preco):

        self.nome = nome

        if preco > 0:

            self.preco = preco

        else:

            self.preco = 0

            print("Preço inválido")


produto = Produto("Monitor", -500)

print(produto.preco)
````

# 🔍 Análise

Agora o próprio construtor protege os dados do objeto logo na criação.

👉 Isso melhora muito a segurança do sistema.
