# 🚀 Criando Nosso Primeiro Módulo


## 📄 Arquivo: produto.py


````python
class Produto:

    def __init__(self, nome, preco):

        self.nome = nome
        self.preco = preco

    def exibir_dados(self):

        print(f"Produto: {self.nome}")
        print(f"Preço: R$ {self.preco}")
````

# 🚀 Importando Módulos


## 📄 Arquivo: app.py


````python
from produto import Produto


produto = Produto("Notebook", 3500)

produto.exibir_dados()
````

# 🔍 O que aconteceu?

Agora o código está separado em arquivos diferentes.

👉 Isso melhora muito a organização.


# 🧱 Modularizando o Sistema Comercial


## 📄 Arquivo: cliente.py


````python
class Cliente:

    def __init__(self, nome, telefone):

        self.nome = nome
        self.telefone = telefone

    def exibir_dados(self):

        print(f"Cliente: {self.nome}")
        print(f"Telefone: {self.telefone}")
````

## 📄 Arquivo: pedido.py


````python
class Pedido:

    def __init__(self, cliente):

        self.cliente = cliente
        self.produtos = []

    def adicionar_produto(self, produto):

        self.produtos.append(produto)
````

## 📄 Arquivo: app.py


````python
from produto import Produto
from cliente import Cliente
from pedido import Pedido


cliente = Cliente("Carlos", "199999999")

produto = Produto("Notebook", 3500)

pedido = Pedido(cliente)

pedido.adicionar_produto(produto)

print("Pedido criado com sucesso!")
````

# 🔥 O que acabamos de fazer?

Agora o sistema possui:

✅ múltiplos arquivos  
✅ organização profissional  
✅ separação de responsabilidades  
✅ arquitetura mais limpa  


# ⚙️ Criando Funções Auxiliares


## 📄 Arquivo: validacoes.py


````python
def validar_preco(preco):

    if preco <= 0:

        return False

    return True
````

## 📄 Utilizando a Função


````python
from validacoes import validar_preco


preco = 3500

if validar_preco(preco):

    print("Preço válido")

else:

    print("Preço inválido")
````
