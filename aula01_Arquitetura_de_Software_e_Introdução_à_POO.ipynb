# 🧠 Contexto

Na programação tradicional é comum desenvolver sistemas utilizando apenas:

- variáveis
- listas
- condicionais
- estruturas de repetição
- funções

Inicialmente isso parece suficiente.

Entretanto, quando os sistemas crescem, começam a surgir diversos problemas:

❌ código duplicado  
❌ dificuldade de manutenção  
❌ bugs frequentes  
❌ dificuldade de escalabilidade  
❌ baixa reutilização  

👉 É exatamente nesse cenário que surge a necessidade da Programação Orientada a Objetos.


# ⚠️ Problema Inicial

Imagine que precisamos desenvolver um sistema para uma loja.

Esse sistema deve permitir:

- cadastrar produtos
- listar produtos
- buscar produtos
- atualizar estoque
- calcular valor total

Vamos começar utilizando uma abordagem procedural.


````python
produtos = []

while True:

    print("\n===== SISTEMA =====")
    print("1 - Adicionar produto")
    print("2 - Listar produtos")
    print("3 - Buscar produto")
    print("4 - Atualizar estoque")
    print("5 - Sair")

    opcao = input("Escolha: ")

    if opcao == "1":

        nome = input("Nome do produto: ")
        preco = float(input("Preço: "))
        estoque = int(input("Estoque: "))

        produto = {
            "nome": nome,
            "preco": preco,
            "estoque": estoque
        }

        produtos.append(produto)

    elif opcao == "2":

        for p in produtos:
            print(p)

    elif opcao == "3":

        busca = input("Buscar produto: ")

        for p in produtos:

            if p["nome"] == busca:
                print("Produto encontrado")
                print(p)

    elif opcao == "4":

        nome = input("Produto: ")

        for p in produtos:

            if p["nome"] == nome:

                novo_estoque = int(input("Novo estoque: "))
                p["estoque"] = novo_estoque

    elif opcao == "5":
        break
````

# 🧬 Introdução à Programação Orientada a Objetos

A Programação Orientada a Objetos organiza o sistema utilizando:

## Objetos

Objetos representam entidades do mundo real.

Exemplos:

- Produto
- Cliente
- Pedido
- Funcionário

Cada objeto possui:

- atributos
- comportamentos


# 📦 Classe vs Objeto

## Classe

A classe funciona como um molde.

Exemplo:

```python
class Produto:
    pass
```

---

## Objeto

O objeto é uma instância criada a partir da classe.

Exemplo:

```python
produto1 = Produto()
```

# 🧠 Adicionando Atributos

````python
class Produto:

    def __init__(self, nome, preco, estoque):

        self.nome = nome
        self.preco = preco
        self.estoque = estoque
````

## self

## atributos

Cada produto possuirá:

# 🚀 Criando Objetos

# 🧱 Adicionando Métodos

Na POO os objetos também possuem comportamentos.


````python
class Produto:

    def __init__(self, nome, preco, estoque):

        self.nome = nome
        self.preco = preco
        self.estoque = estoque

    def exibir_dados(self):

        print(f"Produto: {self.nome}")
        print(f"Preço: R$ {self.preco}")
        print(f"Estoque: {self.estoque}")

    def atualizar_estoque(self, quantidade):

        self.estoque = quantidade
````

# 🚀 Testando os Métodos


````python
produto = Produto("Teclado", 200, 20)

produto.exibir_dados()

produto.atualizar_estoque(50)

print("\nNovo estoque:")
produto.exibir_dados()
````

# 🛒 Aplicando POO no Projeto do Semestre

Agora vamos criar uma primeira versão do sistema utilizando objetos.


````python
class Produto:

    def __init__(self, nome, preco, estoque):

        self.nome = nome
        self.preco = preco
        self.estoque = estoque

    def exibir_dados(self):

        print(f"Produto: {self.nome}")
        print(f"Preço: R$ {self.preco}")
        print(f"Estoque: {self.estoque}")

    def atualizar_estoque(self, quantidade):

        self.estoque = quantidade


produtos = []

while True:

    print("\n===== SISTEMA DE GESTÃO =====")
    print("1 - Cadastrar produto")
    print("2 - Listar produtos")
    print("3 - Atualizar estoque")
    print("4 - Sair")

    opcao = input("Escolha: ")

    if opcao == "1":

        nome = input("Nome: ")
        preco = float(input("Preço: "))
        estoque = int(input("Estoque: "))

        produto = Produto(nome, preco, estoque)

        produtos.append(produto)

    elif opcao == "2":

        for p in produtos:
            p.exibir_dados()

    elif opcao == "3":

        nome = input("Produto: ")

        for p in produtos:

            if p.nome == nome:

                novo = int(input("Novo estoque: "))
                p.atualizar_estoque(novo)

    elif opcao == "4":
        break
````
