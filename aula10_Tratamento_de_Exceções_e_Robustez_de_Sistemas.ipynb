# 🚀 Utilizando try e except


````python
try:

    preco = float(input("Digite o preço: "))

    print(preco)

except:

    print("Valor inválido")
````

# 🔍 Explicação

## try

Bloco onde o código será testado.

---

## except

Executado caso ocorra erro.

👉 Agora o sistema não quebra.


# ⚠️ Tratando Erros Específicos


````python
try:

    numero = int(input("Digite um número: "))

    resultado = 10 / numero

    print(resultado)

except ValueError:

    print("Digite apenas números inteiros")

except ZeroDivisionError:

    print("Não é possível dividir por zero")
````

# 🔥 O que aconteceu?

Agora o sistema consegue identificar:

✅ erro de conversão  
✅ divisão por zero  

👉 Isso deixa o sistema muito mais profissional.


# ⚙️ Utilizando else


````python
try:

    idade = int(input("Digite sua idade: "))

except ValueError:

    print("Idade inválida")

else:

    print(f"Idade cadastrada: {idade}")
````

# 🔍 Explicação

## else

Executado apenas se NÃO ocorrer erro.


# ⚙️ Utilizando finally


````python
try:

    arquivo = open("teste.txt")

except FileNotFoundError:

    print("Arquivo não encontrado")

finally:

    print("Finalizando operação")
````

# 🔍 Explicação

## finally

Executado independentemente de erro.

👉 Muito utilizado para:

- fechar conexões
- finalizar operações
- liberar recursos


# 🛒 Aplicando Tratamento de Exceções no Projeto do Semestre


````python
class Produto:

    def __init__(self, nome, preco, estoque):

        self.nome = nome
        self.preco = preco
        self.estoque = estoque

    def exibir_dados(self):

        print(f"Produto: {self.nome}")
        print(f"Preço: R$ {self.preco}")
        print(f"Estoque: {self.estoque}")
````

# 🚀 Sistema de Cadastro Seguro


````python
produtos = []

while True:

    try:

        nome = input("Nome do produto: ")

        preco = float(input("Preço: "))

        estoque = int(input("Estoque: "))

        produto = Produto(nome, preco, estoque)

        produtos.append(produto)

        print("Produto cadastrado com sucesso!")

    except ValueError:

        print("Erro: digite valores válidos")

    opcao = input("Deseja continuar? (s/n): ")

    if opcao == "n":

        break
````

# 🔍 O que mudou?

Agora o sistema:

✅ não quebra  
✅ trata erros  
✅ orienta o usuário  
✅ continua funcionando  

👉 Isso é essencial em sistemas reais.


# ⚠️ Capturando Exceções como Variáveis


````python
try:

    numero = int(input("Digite um número: "))

except ValueError as erro:

    print("Erro encontrado:")

    print(erro)
````

# 🔍 Explicação

Agora conseguimos acessar informações detalhadas sobre o erro.


# 🚀 Criando Validações Mais Profissionais


````python
while True:

    try:

        preco = float(input("Digite o preço: "))

        if preco <= 0:

            raise ValueError("Preço deve ser maior que zero")

        print("Preço válido")

        break

    except ValueError as erro:

        print(f"Erro: {erro}")
````

# 🔥 O que aconteceu?

Agora conseguimos:

✅ criar nossos próprios erros  
✅ validar regras do sistema  
✅ impedir dados inválidos  


# 🧠 Robustez de Sistemas

Sistemas profissionais precisam:

✅ tratar erros  
✅ validar entradas  
✅ impedir falhas  
✅ continuar funcionando  
✅ orientar o usuário  


# 🛒 Evoluindo o Sistema Comercial


````python
class Produto:

    def __init__(self, nome, preco, estoque):

        if preco <= 0:

            raise ValueError("Preço inválido")

        if estoque < 0:

            raise ValueError("Estoque inválido")

        self.nome = nome
        self.preco = preco
        self.estoque = estoque

    def exibir_dados(self):

        print(f"Produto: {self.nome}")
        print(f"Preço: R$ {self.preco}")
        print(f"Estoque: {self.estoque}")


produtos = []

while True:

    try:

        nome = input("\nNome do produto: ")

        preco = float(input("Preço: "))

        estoque = int(input("Estoque: "))

        produto = Produto(nome, preco, estoque)

        produtos.append(produto)

        print("\nProduto cadastrado com sucesso!")

    except ValueError as erro:

        print(f"\nErro: {erro}")

    opcao = input("\nDeseja cadastrar outro produto? (s/n): ")

    if opcao == "n":

        break


print("\n===== PRODUTOS CADASTRADOS =====")

for p in produtos:

    p.exibir_dados()

    print("----------------")
````
