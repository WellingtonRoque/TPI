# ⚙️ Importando SQLite


````python
import sqlite3
````

# 🚀 Criando Banco de Dados


````python
import sqlite3

conexao = sqlite3.connect("sistema.db")

print("Banco criado com sucesso!")
````

# 🔍 Explicação

## connect()

Cria conexão com o banco.

Se o banco não existir:

👉 o SQLite cria automaticamente.


# 🚀 Criando Cursor


````python
cursor = conexao.cursor()
````

# 🔍 O que é Cursor?

O cursor permite:

✅ executar comandos SQL  
✅ inserir dados  
✅ consultar registros  


# 🚀 Criando Tabela


````python
cursor.execute("""
CREATE TABLE produtos (

    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT,
    preco REAL

)
""")

print("Tabela criada!")
````

# 🔍 Explicação

## CREATE TABLE

Cria tabela no banco.

---

## PRIMARY KEY

Identificador único.

---

## AUTOINCREMENT

Incrementa automaticamente o ID.


# 🚀 Salvando Alterações


````python
conexao.commit()
````

# 🔍 Explicação

## commit()

Salva alterações no banco de dados.


# 🚀 Inserindo Dados


````python
cursor.execute("""
INSERT INTO produtos (nome, preco)
VALUES ('Notebook', 3500)
""")

conexao.commit()

print("Produto cadastrado!")
````

# 🚀 Inserindo Múltiplos Produtos


````python
produtos = [
    ("Mouse", 120),
    ("Teclado", 250),
    ("Monitor", 1500)
]

cursor.executemany("""
INSERT INTO produtos (nome, preco)
VALUES (?, ?)
""", produtos)

conexao.commit()

print("Produtos cadastrados!")
````

# 🔍 Explicação

## executemany()

Permite inserir múltiplos registros.

---

## ?

Placeholder utilizado para segurança e organização.


# 🚀 Consultando Dados


````python
cursor.execute("SELECT * FROM produtos")

dados = cursor.fetchall()

for produto in dados:

    print(produto)
````

# 🔍 Explicação

## SELECT

Consulta dados da tabela.

---

## fetchall()

Retorna todos os registros.


# 🚀 Buscando Um Produto


````python
cursor.execute("""
SELECT * FROM produtos
WHERE nome = 'Mouse'
""")

produto = cursor.fetchone()

print(produto)
````

# 🚀 Atualizando Dados


````python
cursor.execute("""
UPDATE produtos
SET preco = 200
WHERE nome = 'Mouse'
""")

conexao.commit()

print("Produto atualizado!")
````

# 🚀 Excluindo Dados


````python
cursor.execute("""
DELETE FROM produtos
WHERE nome = 'Teclado'
""")

conexao.commit()

print("Produto removido!")
````

# 🛒 Integrando SQLite ao Projeto do Semestre


````python
import sqlite3


class Produto:

    def __init__(self, nome, preco):

        self.nome = nome
        self.preco = preco


conexao = sqlite3.connect("sistema.db")

cursor = conexao.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS produtos (

    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT,
    preco REAL

)
""")

produto = Produto("Notebook", 3500)

cursor.execute("""
INSERT INTO produtos (nome, preco)
VALUES (?, ?)
""", (produto.nome, produto.preco))

conexao.commit()

print("Produto salvo no banco!")
````
