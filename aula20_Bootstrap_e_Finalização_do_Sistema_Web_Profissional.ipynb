# 🧱 Criando Layout Base


## 📄 templates/base.html


````python
<!DOCTYPE html>

<html>

<head>

    <title>Sistema Comercial</title>

    <link
        href="https://cdn.jsdelivr.net/npm/bootstrap@5.3.3/dist/css/bootstrap.min.css"
        rel="stylesheet"
    >

</head>

<body>

    <nav class="navbar navbar-dark bg-dark">

        <div class="container-fluid">

            <span class="navbar-brand">
                Sistema Comercial
            </span>

        </div>

    </nav>

    <div class="container mt-4">

        {% block conteudo %}

        {% endblock %}

    </div>

</body>

</html>
````

# 🔍 O que acabamos de fazer?

Agora criamos:

✅ layout reutilizável  
✅ navbar profissional  
✅ estrutura base  
✅ arquitetura frontend profissional  


# 🌐 Utilizando Herança de Templates


## 📄 templates/index.html


````python
{% extends "base.html" %}

{% block conteudo %}

<h1 class="mb-4">
    Bem-vindo ao Sistema Comercial
</h1>

<a href="/produtos" class="btn btn-primary">
    Ver Produtos
</a>

{% endblock %}
````

# 🎨 Melhorando Página de Produtos


## 📄 templates/produtos.html


````python
{% extends "base.html" %}

{% block conteudo %}

<h1 class="mb-4">
    Produtos
</h1>

<a href="/cadastro" class="btn btn-success mb-3">
    Novo Produto
</a>

<table class="table table-striped table-hover">

    <thead class="table-dark">

        <tr>

            <th>ID</th>
            <th>Nome</th>
            <th>Preço</th>
            <th>Ações</th>

        </tr>

    </thead>

    <tbody>

        {% for produto in produtos %}

        <tr>

            <td>{{ produto[0] }}</td>
            <td>{{ produto[1] }}</td>
            <td>R$ {{ produto[2] }}</td>

            <td>

                <a href="/editar/{{ produto[0] }}"
                   class="btn btn-warning btn-sm">

                    Editar

                </a>

                <a href="/excluir/{{ produto[0] }}"
                   class="btn btn-danger btn-sm">

                    Excluir

                </a>

            </td>

        </tr>

        {% endfor %}

    </tbody>

</table>

{% endblock %}
````

# 🔍 Componentes Bootstrap Utilizados

## table table-striped

---

## btn btn-warning

## btn btn-danger

# 🎨 Melhorando Formulário


## 📄 templates/cadastro_produto.html


````python
{% extends "base.html" %}

{% block conteudo %}

<h1 class="mb-4">
    Cadastro de Produto
</h1>

<form method="POST">

    <div class="mb-3">

        <label class="form-label">
            Nome
        </label>

        <input
            type="text"
            name="nome"
            class="form-control"
        >

    </div>

    <div class="mb-3">

        <label class="form-label">
            Preço
        </label>

        <input
            type="number"
            step="0.01"
            name="preco"
            class="form-control"
        >

    </div>

    <button type="submit"
            class="btn btn-primary">

        Cadastrar

    </button>

</form>

{% endblock %}
````

# 🔥 O que mudou?

Agora o formulário possui:

✅ layout moderno  
✅ campos estilizados  
✅ melhor usabilidade  
✅ aparência profissional  


# 🌐 Melhorando Página de Edição


## 📄 templates/editar_produto.html


````python
{% extends "base.html" %}

{% block conteudo %}

<h1 class="mb-4">
    Editar Produto
</h1>

<form method="POST">

    <div class="mb-3">

        <label class="form-label">
            Nome
        </label>

        <input
            type="text"
            name="nome"
            value="{{ produto[1] }}"
            class="form-control"
        >

    </div>

    <div class="mb-3">

        <label class="form-label">
            Preço
        </label>

        <input
            type="number"
            step="0.01"
            name="preco"
            value="{{ produto[2] }}"
            class="form-control"
        >

    </div>

    <button type="submit"
            class="btn btn-warning">

        Atualizar

    </button>

</form>

{% endblock %}
````
